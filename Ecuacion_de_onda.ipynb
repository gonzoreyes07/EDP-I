```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parámetros
L = 1 # Longitud de la cuerda
c = 1 # Velocidad de la onda
k = 1  # Amplitud
N = 50  # Número de términos
```

```python
# Cálulo del coeficiente de Fourier A_n con la solución analítica obtneida
def A_n(n):
    if n % 2 == 0:
        return 0
    else:
        return (8 / (n * np.pi)**2) * (-1)**((n-1)//2)
```

```python
# Solución u(x, t)
def u(x, t, N_terminos):
    resultado = 0
    for n in range(1, 2*N_terminos + 1, 2):  # Para n impares
        resultado += A_n(n) * np.cos(n * np.pi * c * t / L) * np.sin(n * np.pi * x / L)
    return resultado
```

```python
# Imprimimos la ecuación obtenida
print("Solución analítica de la ecuación de onda:")
print("u(x, t) = sum_{n odd} [ 8 (-1)^((n-1)/2) / (n π)^2 ] sin(n π x) cos(n π t)")
print("\nPrimeros términos no nulos (n = 1, 3, 5):")
for n in [1, 3, 5]:
    coeficiente = A_n(n)
    term = f"{coeficiente:.6f} sin({n} π x) cos({n} π t)"
    print(f"n = {n}: {term}")
```

```python
# Gráfica de la solución
x = np.linspace(0, L, 1000)
tiempo = [0, 0.1, 0.2, 0.3, 0.4]

plt.figure(figsize=(10, 6))
for t in tiempo:
    u1 = [u(xi, t, N) for xi in x]
    plt.plot(x, u1, label=f't = {t:.1f}')

plt.title('Cuerda Vibrante con Deflexión inicial triangular')
plt.xlabel('x')
plt.ylabel('u(x, t)')
plt.legend()
plt.grid(True)
plt.show()
```
