```python
from pylab import *
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definimoa la funcion a trozos.
def f(x):
  if -2 <= x < 0:
    return 0
  elif 0 <= x <= 2:
    return 2 - x
  else:
    return 0
```

```python
def f1(x):
    return np.zeros_like(x)

# Graficamos la primera condición de la función
x1 = np.linspace(-2, 0, 200)
y1 = f1(x1)
plt.plot(x1, y1, label='f(x) = 0 para -2 <= x < 0')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Primera condición inicial de la funcion")
plt.legend()
plt.show()
```

```python
def f2(x):
    return 2 - x

# Valores de x para el intervalo [0, 2]
x2 = np.linspace(0, 2, 200)

# Graficamos la segunda condición de la función
plt.plot(x2, f2(x2), label='f(x) = 2 - x para 0 <= x <= 2')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title('Segunda condición de la función definida a trozos')
plt.legend()
show()
```

```python
def u_0_f1(x, N=20): #N es el numero de armonicos
    f = np.zeros(len(x))
    for n in range(1, N):
        f += (-1)**(n+1) * np.sin(n * np.pi * x) / n
    return (2.0 / np.pi) * f
#Utilizando
f_n_f1 = u_0_f1(x1)
# Graficamos
plt.plot(x1, f_n_f1, label='$f_N(x)$ con $N=20$')
plt.plot(x1, f1(x1), label='$f(x) = 0$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$');
plt.title("Aproximación de la primera condición inicial")
xlim([-2.1, 0.1]), ylim([-1, 1]), grid()
legend()
show()
```

```python
def u_0_f2(x, N=20):
    f = np.zeros(len(x))
    for n in range(1, N):
        f += (-1)**(n+1) * np.sin(n * np.pi * x) / n
    return (2.0 / np.pi) * f
#Aplicamos
f_n_f2 = u_0_f2(x2)
# Graficamos
plt.plot(x2, f_n_f2, label='$f_N(x)$ con $N=20$')
plt.plot(x2, f2(x2), label='$f(x) = 2 - x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$')
plt.title("Aproximación de la segunda condición inicial")
xlim([0, 2.1]), ylim([-1, 3]), grid(), legend()
show()
```

```python
def u_x_t_f1(x, t, N=50):
    f = np.zeros((len(x), len(t)))
    for i in range(len(t)):
        for n in range(1, N):
            f[:, i] += (-1)**(n+1) * np.sin(n * np.pi * x) / n * np.exp(-n**2 * np.pi**2 * t[i])
    return (2.0 / np.pi) * f
x1 = np.linspace(-2, 0, 1000)
t = np.linspace(0, 10, 1000)
# Función U
U_f1 = u_x_t_f1(x1, t)
#Graficando
plt.plot(x1, U_f1[:, 0], label='$u(x,0)$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial (f(x) = 0)")
xlim([-2.01, 0.01]), ylim([-1.01, 1.01]), grid(), legend()
show()
```

```python
def u_x_t_f2(x, t, N=50):
    f = np.zeros((len(x), len(t)))
    for i in range(len(t)):
        for n in range(1, N):
            f[:, i] += (-1)**(n+1) * np.sin(n * np.pi * x) / n * np.exp(-n**2 * np.pi**2 * t[i])
    return (2.0 / np.pi) * f
x2 = np.linspace(0, 2, 1000)
t = np.linspace(0, 10, 1000)
# Función U
U_f2 = u_x_t_f2(x2, t)
# Graficamos
plt.plot(x2, U_f2[:, 0], label='$u(x,0)$')
plt.plot(x2, f2(x2), label='$f(x) = 2 - x$')
plt.xlabel('$x$'); plt.ylabel('$f(x)$')
plt.title("Aproximación de la condición inicial (f(x) = 2 - x)")
xlim([0, 2.01]), ylim([-1.01, 3.01]), grid(), legend()
show()
```
